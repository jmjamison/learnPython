```python
#  install bs4 for python2 and python3-bs4 for python3
# <https://www.crummy.com/software/BeautifulSoup/bs4/doc/>
from bs4 import BeautifulSoup
import requests
import numpy as np
from numpy import array
import pandas as pd
```

```python
url = "http://pokemondb.net/pokedex/all"
```

```python
page = requests.get(url)
```

```python
# NOTE:  read_html produces a **list of datafreames** to get the desired one by index use something[0]
# NOTE2:  read_html needs to read a table
text_list = pd.read_html(page.text)
df = text_list[0]
df
```

```python
print(type(df))
```

```python
f=open('outfile.txt', 'w')
f.write(str(df))
```
